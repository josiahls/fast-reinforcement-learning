## Interpreter Demo

```python
import gym
import numpy as np

from fast_rl.agents.DQN import DQN
from fast_rl.core.Learner import AgentLearnerAlpha
from fast_rl.core.MarkovDecisionProcess import MDPDataBunchAlpha, MDPDatasetAlpha
from fast_rl.core.Interpreter import AgentInterpretationAlpha
%matplotlib inline
    
data = MDPDataBunchAlpha.from_env('CartPole-v1', render='human', bs=64)
model = DQN(data)
learn = AgentLearnerAlpha(data, model)

learn.fit(5)
```

```python
data.to_pickle()
```

```python
data = MDPDataBunchAlpha.from_pickle('CartPole-v1', render='human', bs=64)
model = DQN(data)
learn = AgentLearnerAlpha(data, model)
```

```python
interp = AgentInterpretationAlpha(learn)
interp.plot_q_density()
```

```python
interp.plot_episode(1)
```

```python
interp.plot_rewards_over_iterations(1)
```

```python
interp.plot_rewards_over_iterations(cumulative=True)
```

```python
interp.plot_rewards_over_episodes(cumulative=False)
```

```python
interp.plot_rewards_over_episodes(cumulative=True)
```

```python
interp.iplot_episode(episode=1, fps=1)
interp.current_animation.ipython_display(fps=1, loop=True, autoplay=True)
```
